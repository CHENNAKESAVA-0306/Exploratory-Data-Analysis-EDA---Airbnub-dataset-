```python
# filling the null values by 'price column'

df['price'] = df['price'].fillna(df['price'].median())
```

```python
# dropping the null values by remaining all columns 
df.dropna(inplace=True)
```

```python
# deleting all duplicated rows 
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
# EDA - exploratory data Analysis 
```

```python
# univariate Analysis 

# - identifying the outliers by using boxplot

sns.boxplot(x = 'price',data=df)
```

```python
# removing of outlayers by IQR method 

Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)

print(Q1)
print(Q3)

IQR = Q3-Q1
print(IQR)

lower_bound = Q1-1.5*IQR
upper_bound = Q3-1.5*IQR
print(lower_bound)
print(upper_bound)

df = df[(df['price'] >= lower_bound ) & (df['price'] <= upper_bound)]

```

```python
sns.boxplot(x='price',data=df)
```

```python
# price distribution 
plt.figure(figsize=(8,5))
sns.histplot(data=df,x='price',bins=100)
plt.title("price distribution")
plt.ylabel("Frequency")
plt.show()
```

```python
# price distribution 
plt.figure(figsize=(6,3))
sns.histplot(data=df,x='availability_365')
plt.title("availablity_365 Distribuition")
plt.ylabel("Frequency")
plt.show()
```
